# *Collect the Host Details*

```python
host=[]
for i in coll.find({},{"_id":1, "host":1}):
    host.append(i)
```

```python
HOST=pd.DataFrame(host)
```

HOST DATA

```python
HOST
```

```python
HOST_DATA= {'_id':[],'host_id':[], 'host_url':[], 'host_name':[], 'host_location':[],"host_response_time":[], 'host_thumbnail_url':[], 'host_picture_url':[], 'host_neighbourhood':[], 'host_response_rate':[], 'host_is_superhost':[], 'host_has_profile_pic':[], 'host_identity_verified':[], 'host_listings_count':[], 'host_total_listings_count':[], 'host_verifications':[]}

for i in HOST["_id"]:
    HOST_DATA["_id"].append(i)
for i in HOST["host"]:
    HOST_DATA["host_id"].append(i["host_id"])
    HOST_DATA["host_url"].append(i["host_url"])
    HOST_DATA["host_name"].append(i["host_name"])
    HOST_DATA["host_location"].append(i["host_location"])
    HOST_DATA["host_response_time"].append(i.get("host_response_time"))
    HOST_DATA["host_thumbnail_url"].append(i["host_thumbnail_url"])
    HOST_DATA["host_picture_url"].append(i["host_picture_url"])
    HOST_DATA["host_neighbourhood"].append(i["host_neighbourhood"])
    HOST_DATA["host_response_rate"].append(i.get("host_response_rate"))
    HOST_DATA["host_is_superhost"].append(i["host_is_superhost"])
    HOST_DATA["host_has_profile_pic"].append(i["host_has_profile_pic"])
    HOST_DATA["host_identity_verified"].append(i["host_identity_verified"])
    HOST_DATA["host_listings_count"].append(i["host_listings_count"])
    HOST_DATA["host_total_listings_count"].append(i["host_total_listings_count"])
    HOST_DATA["host_verifications"].append(i["host_verifications"])
```

```python
HOST_DATA
```

```python
HOST_DATAFRAME= pd.DataFrame(HOST_DATA)
```

```python
HOST_DATAFRAME.isnull().sum()
```

Handling the Null Values

```python
HOST_DATAFRAME["host_response_time"].fillna("Not Specified",inplace= True)
HOST_DATAFRAME["host_response_rate"].fillna("Not Specified",inplace= True)
```

```python
HOST_DATAFRAME.isnull().sum()
```

```python
HOST_DATAFRAME
```

```python
EMPTY=[]
for index,row in HOST_DATAFRAME.iterrows():
    if row['host_neighbourhood']=='':
        EMPTY.append(index)
```

```python
EMPTY
```

```python
HOST_DATAFRAME["host_neighbourhood"]=HOST_DATAFRAME["host_neighbourhood"].replace({"":"Not Specified"})
```

Mapping the  Values

```python
HOST_DATAFRAME["host_is_superhost"]=HOST_DATAFRAME["host_is_superhost"].map({False: "No",True: "Yes"})
HOST_DATAFRAME["host_has_profile_pic"]=HOST_DATAFRAME["host_has_profile_pic"].map({False: "No",True: "Yes"})
HOST_DATAFRAME["host_identity_verified"]=HOST_DATAFRAME["host_identity_verified"].map({False: "No",True: "Yes"})
```
